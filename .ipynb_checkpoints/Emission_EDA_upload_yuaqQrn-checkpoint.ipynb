###  Data Preprocessing

Clean and merge datasets

```python
#load data
gdp = pd.read_csv("gdp.csv")
population = pd.read_csv("population.csv")
vehicle = pd.read_csv("vehicle_sales.csv")
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

##### External Datasets

```python
# Define a function to convert quarterly data to monthly data
def process_gdp(df):
    
    # convert column headers to lowercase
    df.columns = df.columns.str.lower()
    
    # duplicate each row three times
    df_mth = df.loc[df.index.repeat(3)]

    # create a dictionary to map each quarter to the corresponding months
    quarter_to_months = {
        'Q1': ['01', '02', '03'],
        'Q2': ['04', '05', '06'],
        'Q3': ['07', '08', '09'],
        'Q4': ['10', '11', '12']
    }

    # create a list of monthly timestamps based on the original quarters
    months = []
    for _, row in df.iterrows():
        months.extend([f"{row['year']}-{m}" for m in quarter_to_months[row['quarter']]])

    # set the 'Month' column to the monthly timestamps
    df_mth['month'] = months
    df_mth['month'] = pd.to_datetime(df_mth['month']).dt.strftime('%Y-%m-%d')

    # forward fill the GDP values for each quarter
    df_mth = df_mth.ffill()

    return df_mth


# Define a function to convert annually data to monthly data
def process_general(df):

    # convert column headers to lowercase
    df.columns = df.columns.str.lower()
    
    # Convert 'Year' column to datetime format
    df['year'] = pd.to_datetime(df['year'], format='%Y')
    
    # Set 'Year' column as index
    df = df.set_index('year')
    
    # Resample the data to get the population on the first day of each month
    df_mth = df.resample('MS').asfreq()
    
    # Forward-fill missing values
    df_mth = df_mth.ffill()
    
    # Add a 'Month' column with the first day of each month
    df_mth['month'] = df_mth.index.strftime('%Y-%m-%d')
    
    # Reset index
    df_mth = df_mth.reset_index()
    
    del df_mth['year']
    
    return df_mth
```

```python
# Process dataframes
df_gdp = process_gdp(gdp)
df_pop = process_general(population)
df_veh = process_general(vehicle)
```

##### Train dataset

```python
def inter_extra_polate(df,
                column_name = "month",
                by          = "MS",
                start       = "1987-01-01 00:00:00",
                end         = "2021-12-01 00:00:00"):
    
    """
    this function standardize the time range from "arg:start" to "arg:end" with linear interpolation and extrapolation.
    """
    
    # set the index of the dataframe to the timestamps
    df.set_index(column_name, inplace=True)

    # generate a series of timestamps at 1-m intervals between start_time and end_time
    extrap_index = pd.date_range(start = pd.to_datetime(start),
                                 end   = pd.to_datetime(end),
                                 freq  = by)

    # reindex the dataframe with the new timestamps, filling missing values with NaN
    df = df.reindex(extrap_index)
    
    # fill missing values with linear interpolation between neighboring values
    df = df.interpolate(method='linear', limit_direction="both").reset_index()
    
    # rename index to "time"
    df[column_name] = df["index"]
    del df["index"]
    
    return df
```

```python
# change time resolution from year-month to datetime
train["month"] = train.apply(lambda row : pd.to_datetime(str(row["year"]) + " " + str(row["month"])), axis = 1)

# delete unnecessary columns
del train["year"]

# merge  source and fueltype to source_fueltype
train["source_fueltype"] = train[["source", "fuel_type"]].apply("-".join, axis=1)
del train["source"]
del train["fuel_type"]

# unmelt
train = train.pivot(index='month', columns='source_fueltype', values='emissions_tons') 
train = train.reset_index()

# apply interpolation
train = inter_extra_polate(train).reset_index(drop=True)

# convert month
train['month'] = train['month'].dt.strftime('%Y-%m-%d')
```

##### Merge Datasets

```python
# Merge the data frames on 'month' column
df_merged = pd.merge(train, df_gdp, on='month', how='left')
df_merged = pd.merge(df_merged, df_pop, on='month', how='left')
df_merged = pd.merge(df_merged, df_veh, on='month', how='left')
```

```python
# convert 'month' column to datetime type
df_merged['month'] = pd.to_datetime(df_merged['month'])

# filter the dataframe to select rows where month is on or after 1993-01-01
df_filtered = df_merged[df_merged['month'] >= '1993-01-01']

# remove unrequired columns 
del df_filtered['year']
del df_filtered['quarter']

df_filtered.head()
```

### Feature Selection

Explore relationships between variables and create new features

```python
df = df_filtered.copy()
```

##### 1. Correlation plot - independent variables

Evaluate relationship between independent variables

```python
# Select relevant columns
cols = ['gdp_(current_mkt_value)', 'gdp_(ref_year=2002)', 'population',
       'passenger', 'commercial', 'motorcycle', 'total']


# Compute correlation matrix
corr = df[cols].corr()

# Plot heatmap
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='coolwarm', annot=True, linewidths=.5, ax=ax)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

##### 2. Feature extraction

Based on the correlation plot above, we can remove features with duplicated information and create new features to account for strong correlations (>0.7) between independent variables. 

```python
# 2a. GPD per capital - account for relationship between GDP and population growth
df['gdp_cap'] = df['gdp_(current_mkt_value)']/df['population']

# 2b. proportion of vehicles
df['pro_passenger'] = df['passenger']/df['total']
df['pro_commercial'] = df['commercial']/df['total']
df['pro_motorcycle'] = df['motorcycle']/df['total']
```

```python
# remove duplicated features
columns_to_drop = ['gdp_(current_mkt_value)', 'gdp_(ref_year=2002)', 'population', 
                   'passenger', 'commercial', 'motorcycle', 'total']
df.drop(columns_to_drop, axis=1, inplace=True)
```

##### 3. Normalise variable values 

```python
# select only columns with float data type
cols_to_normalize = df.select_dtypes(include=[np.float]).columns.tolist()

# initialize the scaler
scaler = MinMaxScaler()

# fit and transform the data
df[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])
```

##### 2. Decomposition

One way to find similar dependent variables in terms of seasonality and trend index is to look for variables with similar values across these indices. For example, variables with similar values for their seasonal index can be considered to have similar seasonality patterns. Similarly, variables with similar values for their trend index can be considered to have similar trend patterns.

Here we compare Group A variables vs Group B variables

```python
# set the index as the date column month convert to datetime format
df['month'] = pd.to_datetime(df['month'])
df.set_index('month', inplace=True)

vars_to_compare = ['industry-coal', 'industry-natural_gas', 'transport-natural_gas', 'transport-oil']

for var in vars_to_compare:
    # perform time series decomposition on the variable
    decomposition = sm.tsa.seasonal_decompose(df[var], model='additive')

    # plot the trend and seasonal components of the variable and industry-oil and other-oil
    fig, ax = plt.subplots(2, 1, figsize=(12, 8))
    fig.suptitle(var)
    
    # plot the trends
    decomposition.trend.plot(ax=ax[0], label=var + ' Trend')
    df['industry-oil'].plot(ax=ax[0], label='industry-oil Trend')
    df['other-oil'].plot(ax=ax[0], label='other-oil Trend')
    ax[0].set_title('Trend Comparison')
    ax[0].legend()
    
    # plot the seasonality
    decomposition.seasonal.plot(ax=ax[1], label=var + ' Seasonality')
    sm.tsa.seasonal_decompose(df['industry-oil'], model='additive').seasonal.plot(ax=ax[1], label='industry-oil Seasonality')
    sm.tsa.seasonal_decompose(df['other-oil'], model='additive').seasonal.plot(ax=ax[1], label='other-oil Seasonality')
    ax[1].set_title('Seasonality Comparison')
    ax[1].legend()

    plt.tight_layout()
    plt.show()
```

Generate a table displaying the seasonality and trend indices. Based on the analysis, the variables with the highest similarity in terms of seasonality and trend are:

- Industry oil with transport oil
- Other oil with transport natural gas

To improve the model performance of industry oil and other oil, we can consider applying the models developed for transport oil and transport natural gas, respectively. This approach takes advantage of the similarity in seasonality and trend between the variables and can lead to better forecast accuracy.

```python
# initialize empty list to store similar variables
similar_vars = []

# loop through vars_to_compare list
for var in vars_to_compare:
    # perform time series decomposition for both industry-oil and the current variable
    decomp1 = sm.tsa.seasonal_decompose(df['industry-oil'], model='additive')
    decomp2 = sm.tsa.seasonal_decompose(df[var], model='additive')

    # compare trends and seasonality using correlation
    corr_trend = decomp1.trend.corr(decomp2.trend)
    corr_seasonal = decomp1.seasonal.corr(decomp2.seasonal)

    # if both correlations are above a certain threshold, consider variables as similar
    if corr_trend > 0.6 and corr_seasonal > 0.6:
        similar_vars.append(('industry-oil', var))

    # perform time series decomposition for both other-oil and the current variable
    decomp1 = sm.tsa.seasonal_decompose(df['other-oil'], model='additive')
    decomp2 = sm.tsa.seasonal_decompose(df[var], model='additive')

    # compare trends and seasonality using correlation
    corr_trend = decomp1.trend.corr(decomp2.trend)
    corr_seasonal = decomp1.seasonal.corr(decomp2.seasonal)

    # if both correlations are above a certain threshold, consider variables as similar
    if corr_trend > 0.6 and corr_seasonal > 0.6:
        similar_vars.append(('other-oil', var))

# print list of similar variables
print('Similar variables:')
for pair in similar_vars:
    print(pair)

# generate table of correlations between trends and seasonality
corr_df = pd.DataFrame(index=['industry-oil', 'other-oil']+vars_to_compare, columns=['industry-oil', 'other-oil']+vars_to_compare)
for i in range(len(dep_vars)):
    for j in range(i+1, len(dep_vars)):
        var1 = dep_vars[i]
        var2 = dep_vars[j]

        # perform time series decomposition for both variables
        decomp1 = sm.tsa.seasonal_decompose(df[var1], model='additive')
        decomp2 = sm.tsa.seasonal_decompose(df[var2], model='additive')

        # compute correlations between trends and seasonality
        corr_trend = decomp1.trend.corr(decomp2.trend)
        corr_seasonal = decomp1.seasonal.corr(decomp2.seasonal)

        # store correlations in table
        corr_df.loc[var1, var2] = '{:.2f}/{:.2f}'.format(corr_trend, corr_seasonal)
        corr_df.loc[var2, var1] = '{:.2f}/{:.2f}'.format(corr_trend, corr_seasonal)

# display table
print('\nCorrelation between trends and seasonality:')
display(corr_df.iloc[:, :2])
```

### Summary

In conclusion, we have clean and merge the dependent and independent variables from various datasets. We have also applied feature selection techniques to identify strong correlated dependent and independent variables and account for correlations between similar features. However, for dependent variables without strongly correlated independent variables, the forecast accuracy may not compromised when using a multivariate forecast technique. In such cases, we can overcome this limitation by using univariate techniques or applying a cluster approach, where models from other dependent variables that share similar seasonality and trend can be used. This approach can improve the accuracy of the forecasted output and help to make better-informed decisions based on the data.
